```python
import numpy as np
import matplotlib.pyplot as plt

N = [1, 2, 3, 4, 5, 6, 7, 8, 9]
errors = [11.4283, 7.44047, 2.13609, 0.285627, 0.0223289, 0.00114945, 4.20069e-05, 1.14786e-06, 2.43612e-08]
plt.grid()
plt.plot(N, errors)
plt.xlabel(r'N', loc = 'right')
plt.ylabel(r'error', rotation = 0, loc = 'top')
plt.show()
```
