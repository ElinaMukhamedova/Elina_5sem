```python
import numpy as np
import matplotlib.pyplot as plt
log10_error1 = [-1.94311, -5.27552, -11.9428]
log10_error2 = [-1.46955, -3.97144, -8.97934]
log10_error3 = [-0.50991, -1.62534, -3.80581, -8.16314]
log10_error4 = [0.369527, -1.58247, -3.55882, -7.52074, -15.3525]
plt.grid()
plt.plot([1, 2, 3], log10_error1, label = "ecc = 0.1")
plt.plot([1, 2, 3], log10_error2, label = "ecc = 0.2")
plt.plot([1, 2, 3, 4], log10_error3, label = "ecc = 0.5")
plt.plot([1, 2, 3, 4, 5], log10_error4, label = "ecc = 0.8")
plt.xlabel(r'iteration', loc = 'right')
plt.ylabel(r'log10(error)', rotation = 0, loc = 'top')
plt.legend()
plt.show()
```
